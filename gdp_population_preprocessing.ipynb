## Given 2013 contains least not null values and is one of the latest year, we will consider this as our metric

```python
countries = countries[['Country Name','Country Code','2013']]
countries.head(n=5)
```

```python
countries['2013'] = countries['2013'].fillna(0.00).apply(np.int64)
countries.rename(columns={'2013' : 'GDP per capita (current US$)'}, inplace=True)
countries.head(n=5)
```

```python
population = pd.read_csv('population_density.csv')
population.head(n=5)
```

```python
population = population[['Country Name','Country Code','2013']]
population['2013'] = population['2013'].fillna(0.00).apply(np.int64)
population.rename(columns={'2013' : 'Population density'}, inplace=True)
population.head(n=5)
```

```python
countries
```

```python
population
```

```python
merged_df = pd.merge(countries, population, left_on=['Country Code','Country Name'], right_on=['Country Code','Country Name'])
merged_df
```

```python
merged_df.to_csv('gdp_per_capita_and_population_density.csv')
```
